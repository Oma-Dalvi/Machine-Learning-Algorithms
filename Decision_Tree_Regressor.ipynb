```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import datasets,metrics
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
diabetes = datasets.load_diabetes()
diabetes.keys()
```

```python
# Independent Variable
diabetes.feature_names
```

```python
# Dependent Variable
diabetes.target
```

```python
x = diabetes.data
y = diabetes.target

x.shape, y.shape
```

```python
df = pd.DataFrame(x, columns=diabetes.feature_names)
df["target"] = y
df.head()
```

```python
sns.pairplot(df)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y , test_size = 0.2,
                                                 random_state = 42)
```

```python
regressor = DecisionTreeRegressor(random_state=42)
regressor.fit(x_train,y_train)
y_pred = regressor.predict(x_test)
```

```python
y_pred
```

```python
plt.figure(figsize=(16,2))
plt.plot(y_pred)
plt.plot(y_test) # Actual-value Or True_value
```

```python
#MSE
np.sqrt(metrics.mean_squared_error(y_test,y_pred))
```

```python
# Standard_deviation
y_test.std()
```

#### MSE is less than the standard_deviation which means that model is good

### Theory

### Decision Tree Regressor
##### Used To predict target variables in (Continuous form) Non Linear Passion
##### Decision Tree Regression is non linear Regression.
#### Decision Tree Regression is used to predict the target variable whose values are Continuous in nature.(Like 4.1,4.2,2.2 etc)

###### 1] sort all the the values in ascending order    2] set Threshold values,Depending upon threshold spliting will take place     3] Again set new Threshold Values make split till we get best Entropy along with Highest information Gain.

###### check Better Entropy model along with the Information Gain,Whichever having Highest Information Gain it'll be selecting that particular Branch.

#### Disadvantage :- As if Large Dataset gets then Time Complexity is Main Isssue in Decision Tree Regressor.
